```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from cnugs_ml.data import describe_array, normalize_images, ensure_channel_dim, prepare_images
from cnugs_ml.plots import plot_image_grid, set_plot_style

set_plot_style()
```

## Load your data

Replace this example with the hackathon dataset. The fallback uses MNIST so the notebook runs immediately.

```python
from tensorflow.keras.datasets import mnist

(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

describe_array("train_images raw", train_images)
describe_array("train_labels", train_labels)
describe_array("test_images raw", test_images)
```

## Normalize and fix shape

```python
x_train = prepare_images(train_images, add_channel=True)
x_test = prepare_images(test_images, add_channel=True)

describe_array("x_train prepared", x_train)
describe_array("x_test prepared", x_test)
```

## Visual sanity check

```python
plot_image_grid(x_train, train_labels, n=12, rows=3)
```

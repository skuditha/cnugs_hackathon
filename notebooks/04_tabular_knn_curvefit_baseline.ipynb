```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

from cnugs_ml.classical import load_iris_dataframe, train_knn_classifier, sweep_k_neighbors
from cnugs_ml.regression import sine_model, fit_curve, chi2_ndf
from cnugs_ml.plots import set_plot_style

set_plot_style()
```

## Curve fit example

```python
x = np.linspace(0, 2*np.pi, 100)
y = 1.2*np.sin(x) + 0.1*np.random.randn(100)

popt, pcov, perr = fit_curve(sine_model, x, y)
y_fit = sine_model(x, *popt)

print(f"a = {popt[0]:.3f} ± {perr[0]:.3f}")
print(f"b = {popt[1]:.3f} ± {perr[1]:.3f}")
print(f"chi2/ndf = {chi2_ndf(y, y_fit, len(popt)):.3f}")

plt.plot(x, y, "o", label="data")
plt.plot(x, y_fit, label="fit")
plt.legend()
plt.show()
```

## kNN classifier example

```python
df = load_iris_dataframe()
df.head()
```

```python
x_data = df.drop("type", axis=1)
y_data = df["type"]

x_train, x_val, y_train, y_val = train_test_split(
    x_data, y_data, test_size=0.2, random_state=27, stratify=y_data
)

knn = train_knn_classifier(x_train, y_train, n_neighbors=3)
print(f"validation accuracy = {knn.score(x_val, y_val):.3f}")
```

```python
scores = sweep_k_neighbors(x_train, y_train, x_val, y_val, k_values=range(1, 21))
plt.plot(scores["n_neighbors"], scores["accuracy"], marker="o")
plt.xlabel("Number of neighbors")
plt.ylabel("Validation accuracy")
plt.show()
scores.sort_values("accuracy", ascending=False).head()
```
